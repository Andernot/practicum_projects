# Описание данных

Данные состоят из нескольких файлов, полученных из разных источников:

**data_arc_new.csv** — данные об электродах, где

- `key` — номер партии;
- `Начало нагрева дугой` — время начала нагрева;
- `Конец нагрева дугой` — время окончания нагрева;
- `Активная мощность` — значение активной мощности;
- `Реактивная мощность` — значение реактивной мощности,

**data_bulk_new.csv** — данные о подаче сыпучих материалов (объём), где

- `key` — номер партии;
- `Bulk 1 … Bulk 15` — объём подаваемого материала,

**data_bulk_time_new.csv** — данные о подаче сыпучих материалов (время), где

- `key` — номер партии;
- `Bulk 1 … Bulk 15` — время подачи материала,

**data_gas_new.csv** — данные о продувке сплава газом, где

- `key` — номер партии;
- `Газ 1` — объём подаваемого газа,

**data_temp_new.csv** — результаты измерения температуры, где

- `key` — номер партии;
- `Время замера` — время замера;
- `Температура` — значение температуры,

**data_wire_new.csv** — данные о проволочных материалах (объём), где

- `key` — номер партии;
- `Wire 1 … Wire 9` — объём подаваемых проволочных материалов,

**data_wire_time_new.csv** — данные о проволочных материалах (время), где

- `key` — номер партии;
- `Wire 1 … Wire 9` — время подачи проволочных материалов.

Во всех файлах столбец key содержит номер партии. В файлах может быть несколько строк с одинаковым значением key: они соответствуют разным итерациям обработки.

### Посмотрим как взаимодействуют признаки между собой в объединенной таблице. Создадим матрицу корреляций.

```python
plt.figure(figsize=(20, 10));
sns.heatmap(df.corr(), annot=True, fmt='.1f');
```

Наибольшая корреляция у целевого признака наблюдается с признаками Bulk 12, Bulk 15, Wire 1. Для признака Wire 5 нет данных. Также присутствуют мультиколлинеарные признаки: 1)`Bulk 9` и `Wire 8`; 2) `Активная мощность`, `Реактивная мощность` и `Полная мощность`.

Удалим столбец, в котором нет информации `Wire 5`. Из мультиколлинеарных признаков удалим колонки `Wire 8`, `Активная мощность`, `Реактивная мощность`. Также предположим что номер плавки не влияет на технологический процесс и удалим столбец `key`.

```python
df = df.drop(['Wire 8', 'Wire 5', 'key', 'Активная мощность', 'Реактивная мощность'], axis=1)
```

Для обучения моделей отберем признаки, которые сильнее коррелируют с целевым признаком `Температура завершения`.

```python
def important_col(data):
    important_col = []
    for i in df.columns:
        if abs(df['Температура завершения'].corr(df[i])) >= 0.05:
            important_col.append(i)
    data = data[important_col]
    return data
```

## Обучение моделей машинного обучения

### Модель линейной регрессии LinearRegression

```python
%%time
model = LinearRegression()
mae = cross_val_score(model, features_train, targets_train, 
                      scoring='neg_mean_absolute_error', cv=3, n_jobs=-1).mean()

print("MAE на кросс-валидации:", round(abs(mae), 2))
```

### Модель градиентного бустинга LGBMRegressor

```python
%%time
model_lgbm = LGBMRegressor(random_state=10724, n_jobs=-1, verbose=-1)
prm = {'max_depth': [x for x in range(4, 6)], 'n_estimators': [x for x in range(65, 80)], 
      'learning_rate': [x for x in np.arange(0.08, 0.11, 0.01)]}
lgbmgrid = GridSearchCV(estimator=model_lgbm, cv=3, 
                        param_grid=prm, 
                        scoring='neg_mean_absolute_error')
params_model = lgbmgrid.fit(features_train, targets_train)
print(f'Лучшие гиперпараметры модели LGBMRegressor: {params_model.best_estimator_}.')
print(f'Среднее значение MAE на кросс-валидации для модели LGBMRegressor равно: \
{round(abs(params_model.best_score_), 2)}.')
```

### Модель градиентного бустинга CatBoostRegressor

```python
%%time
model_ctbst = CatBoostRegressor(random_state=10724, verbose=-1)
ctbstgrid = GridSearchCV(estimator=model_ctbst, cv=3, 
                        param_grid={
                            'max_depth': [x for x in range(5, 8)], 
                            'iterations': [x for x in range(110, 131, 2)], 
                            'learning_rate': [x for x in np.arange(0.08, 0.13, 0.01)]},  
                        scoring='neg_mean_absolute_error')
params_ctbstmodel = ctbstgrid.fit(features_train, targets_train, verbose=0)
print(f'Лучшие гиперпараметры модели CatBoostRegressor: {params_ctbstmodel.best_params_}.')
print(f'Среднее значение MAE на кросс-валидации для модели CatBoostRegressor равно: \
{round(abs(params_ctbstmodel.best_score_), 2)}.')
```

Лучшее значение МАЕ (6.15) на кросс-валидации показала модель CatBoostRegressor: `{'iterations': 128, 'learning_rate': 0.11, 'max_depth': 6}`

## Сравнение лучшей модели с моделью DummyRegressor

```python
dummy_model = DummyRegressor(strategy='mean')

dummy_mae = cross_val_score(dummy_model, features_train, targets_train, 
                      scoring='neg_mean_absolute_error', cv=3, n_jobs=-1).mean()

print("MAE на кросс-валидации для DummyRegressor со стратегией 'mean':", round(abs(dummy_mae), 2))
```

Наша лучшая модель показала значение MAE на кросс-валидации лучше чем модель `DummyRegressor(strategy='mean')`.

## Выбор лучшей модели

```python
model_ctbst = CatBoostRegressor(
    random_state=10724, max_depth=6, iterations=128, learning_rate=0.11)
model_ctbst.fit(features_train, targets_train, verbose=100)
pred_test = model_ctbst.predict(features_test)
mae = mean_absolute_error(targets_test, pred_test)
print(f'Значение MAE на тестовой выборке для лучшей модели равно: {round(mae, 2)}')
```
